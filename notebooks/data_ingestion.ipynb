## Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/data_stocks.csv")
```

```python
df.head()
```

```python
df.info()
```

## Data Preparation

```python
df['DATE'] = pd.to_datetime(df['DATE'],unit='s')
```

```python
df.set_index('DATE',inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[['NASDAQ.AAPL','NASDAQ.ADP','NASDAQ.CBOE','NASDAQ.CSCO','NASDAQ.EBAY']]
```

```python
df.head()
```

```python
df.to_csv('../data/data.csv')
```

## Data Visualization

```python
df.plot(figsize=(12,4))
plt.ylabel('NASDAQ stocks')
plt.title("Stocks")
plt.xlabel('Time')
plt.show()
```
